```python
import os
import pandas as pd
import re

# Function to extract search keyword and date from filename
def extract_details_from_filename(filename):
    parts = filename.split('_')
    search_keyword = '_'.join(parts[2:-1]).replace('.csv', '').replace('_', ' ')
    date_match = re.search(r'\d{4}-\d{2}-\d{2}', filename)
    date = date_match.group(0) if date_match else None
    return search_keyword, date

# Function to concatenate all CSV files with extracted details
def concatenate_csv_files(folder_path):
    dataframes = []
    for file in os.listdir(folder_path):
        if file.endswith('.csv'):
            file_path = os.path.join(folder_path, file)
            search_keyword, date = extract_details_from_filename(file)
            df = pd.read_csv(file_path)
            df['Search Keyword'] = search_keyword
            df['Date'] = date
            dataframes.append(df)
    combined_df = pd.concat(dataframes, ignore_index=True)
    return combined_df

# Function to match keywords in the 'description' column
def match_keywords(df, tools, industry_skills):
    lowercase_keywords = [keyword.lower() for keyword in tools + industry_skills]
    regex_pattern = r'\b(?:' + '|'.join(re.escape(keyword.strip()) for keyword in lowercase_keywords) + r')\b'
    df['Match Keywords'] = df['description'].apply(
        lambda x: ', '.join(sorted(set(re.findall(regex_pattern, x.lower())))) if pd.notnull(x) else ''
    )
    return df

# Function to count the number of matching keywords
def count_keywords(df):
    df['Number of Matching Keywords'] = df['Match Keywords'].apply(
        lambda x: len(x.split(', ')) if x else 0
    )
    return df

# Main function to run the entire process
if __name__ == "__main__":
    # Define folder path and keywords
    folder_path = r"C:\Users\Richard\Documents\We Cloud Data\Business Development\public biz dev\biz_dev_public\Full workflow Indeed\job_details_Indeed"
    tools = ["AWS", "MS Access", "Microsoft Access", "Azure", " C ", " C,", "C++", "Cassandra", "CircleCI", "Cloud", "Confluence", "Databricks", "Docker", "EMR", "ElasticSearch",
        " Excel ", "Flask", "MLFlow", "Kubeflow", "GCP", " Git ", "Github", "Hadoop", "Hive", "Hugging Face", "Informatica", "JIRA", "Java", "Javascript",
        "Jenkins", "Kafka", "Keras", "Kubernetes", "LLMs", "Matlab", "Mongodb", "MySQL", "New Relic", "NoSQL", "Numpy", "Oracle",
        "Pandas", "PostgreSQL", "Postman", "Power BI", "PySpark", "Python", "Pytorch", "Quicksight", " R ", " R, ", "Redshift", "S3",
        "SAP", "SAS","SQL", "SQL Server", "Scala", "Scikit-learn", "Snowflake", "Spacy", "Spark", "StreamLit", "Tableau",
        "Talend", "Tensorflow", "Terraform", "Torch", "VBA", "XML", "transformer", "CI/CD"]
    industry_skills = ["API Design", "API Development", "Batch Processing", "Big data", "Bioinformatics", "Business Intelligence", "CI/CD",
                "Classification", "Cloud", "Cloud Computing", "Containerization", "Critical Thinking", "Data Analysis",
                "Data Architecture", "Data Cleaning", "Data Extraction", "Data Governance", "Data Ingestion", "Data Integration",
                "Data Manipulation", "Data Mining", "Data Modeling", "Data Pipelines", "Data Security", "Data Visualization",
                "Data Warehousing", "Data Wrangling", "Database Design", "Deep Learning", "DevOps", "Distributed computing", "ETL",
                "Econometrics", "Extract", "Feature Engineering", "Google Cloud", "Kubernetes", "LLMs", "Load (ETL) Processes",
                "Logging", "ML", "Machine Learning", "Mathematics", "Metrics", "Microservices Architecture", "Model Deployment",
                "Model Monitoring", "NLP", "Natural Language Processing", "Natural Language Understanding",
                "Report Generation", "Research Skills",
                "Scripting", "Statistical Analysis", "Statistics", "Technical Documentation", 
                "Understanding of Machine Learning Algorithms"]

    # Process the CSV files
    combined_df = concatenate_csv_files(folder_path)
    combined_df = match_keywords(combined_df, tools, industry_skills)
    combined_df = count_keywords(combined_df)

    column_order= ['Date','Search Keyword','Match Keywords','Number of Matching Keywords','name', 'key', 'title', 'location', 'jobType', 'posted', 'days_ago',
       'rating', 'experience', 'salary', 'education', 'feed', 'link', 'Tools',
       'Soft Skills', 'Industry Skills', 'description' ]

    # Save the final DataFrame to a CSV file
    output_path = os.path.join('final_combined_job_details.csv')
    combined_df[column_order].to_csv(output_path, index=False)
    print(f"Final combined CSV saved to: {output_path}")
```
